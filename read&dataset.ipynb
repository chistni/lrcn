```python
# Library
import torch
from torch import nn
import torch.utils.data as Data
from d2l import torch as d2l
import pandas

from PIL import Image
import numpy as np
import os
```

```python
# 读取第一张图片
img = Image.open(r'.\ba2D_tiffs_temperature\ba2D_28400_T.tif')
img = img.crop((200,415,1020,765))
x_image = np.array(img)
print(x_image.shape)
x_image = np.expand_dims(x_image, axis=0)
print(x_image.shape)
```

```python
# 批量读取图片
for dirpath, dirname, filenames in os.walk(r'.\ba2D_tiffs_temperature'):
    for filename in filenames:
        if (filename != 'Thumbs.db') & (filename != 'ba2D_28400_T.tif'):
            im = Image.open(dirpath + '\\' + filename)
            im = im.crop((200,415,1020,765))
            image_array = np.array(im)
            image_array = np.expand_dims(image_array, axis=0)
            x_image = np.concatenate((x_image, image_array), axis=0)
print(x_image.shape)
```

```python
filenames = np.delete(filenames, [120])
#print(filenames)
```

```python
# filetime(y_train)预处理
filetime = np.empty((120))
for i in range(120):
    if (28400 + 50*i) <= 32500:
        filetime[i] = 28400 + 50*i
    else:
        filetime[i] = 28400 + 50*(i + 1)
#print(filetime)
```

```python
# 将x_image和y_time转换为torch.tensor
x_image = torch.tensor(x_image, dtype=torch.float32)
print(type(x_image), x_image.shape)
y_time = torch.tensor(filetime, dtype=torch.int32)
print(type(y_time), y_time.shape)
```

```python
# 数据集归一化
x_image = x_image/255
```

```python
# 创建TensorDataset对象并划分数据集和训练集
train_dataset = Data.TensorDataset(x_image[0:100,:,:], y_time[0:100])
test_dataset = Data.TensorDataset(x_image[100:120,:,:], y_time[100:120])
print(type(train_dataset), type(test_dataset))
```

```python
# 使用DataLoader对数据集进行批处理

batch_size = 10

train_loader = Data.DataLoader(
    dataset = train_dataset,
    batch_size = batch_size,
    shuffle=False
)
test_loader = Data.DataLoader(
    dataset = test_dataset,
    batch_size = batch_size,
    shuffle=False
)

n_epoch = 1
for epoch in range(n_epoch):
    i = 0
    for x_train, y_train in train_loader:
        i = i + 1
        print('Epoch:{} | num:{} | x_train:{} | y_train:{}'.format(epoch, i ,x_train, y_train))
    i = 0
    for x_test, y_test in test_loader:
        i = i + 1
        print('Epoch:{} | num:{} | x_test:{} | y_test:{}'.format(epoch, i ,x_test, y_test))    
```
